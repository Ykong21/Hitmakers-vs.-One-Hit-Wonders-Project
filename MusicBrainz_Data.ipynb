```python
####This is the function to grab Primary Key Table########
import os
import pandas as pd

file_path = "artist/mbdump/artist"
out_csv = "artists_flat_full.csv"

def flatten_artists(df: pd.DataFrame) -> pd.DataFrame:
    out = pd.DataFrame({
        "mbid": df.get("id"),
        "name": df.get("name"),
        "sort_name": df.get("sort-name"),
        "disambiguation": df.get("disambiguation"),
        "type": df.get("type"),
        "type_id": df.get("type-id"),
        "country": df.get("country"),
        "gender": df.get("gender"),          # optional
        "gender_id": df.get("gender-id"),    # optional
    })

    # life-span dict
    ls = df.get("life-span")
    if ls is not None:
        out["begin_date"] = ls.apply(lambda x: x.get("begin") if isinstance(x, dict) else None)
        out["end_date"]   = ls.apply(lambda x: x.get("end") if isinstance(x, dict) else None)
        out["ended"]      = ls.apply(lambda x: x.get("ended") if isinstance(x, dict) else None)
    else:
        out["begin_date"] = None
        out["end_date"] = None
        out["ended"] = None

    # area dict
    area = df.get("area")
    if area is not None:
        out["area_id"]   = area.apply(lambda x: x.get("id") if isinstance(x, dict) else None)
        out["area_name"] = area.apply(lambda x: x.get("name") if isinstance(x, dict) else None)
    else:
        out["area_id"] = None
        out["area_name"] = None

    # rating dict
    rating = df.get("rating")
    if rating is not None:
        out["rating_value"] = rating.apply(lambda x: x.get("value") if isinstance(x, dict) else None)
        out["rating_votes"] = rating.apply(lambda x: x.get("votes-count") if isinstance(x, dict) else None)
    else:
        out["rating_value"] = None
        out["rating_votes"] = None

    return out

chunk_size = 200_000
first = True
for df_chunk in pd.read_json(file_path, lines=True, chunksize=chunk_size):
    flat = flatten_artists(df_chunk)
    flat.to_csv(out_csv, index=False, mode="w" if first else "a", header=first)
    first = False

print("done:", out_csv)
```

```python
import os
import pandas as pd

file_path = "artist/mbdump/artist"
out_dir = "artist_child_csv"  # I made a separate folder for extra CSVs. feel free to delete it
os.makedirs(out_dir, exist_ok=True)

aliases_csv = os.path.join(out_dir, "artist_aliases.csv")
tags_csv    = os.path.join(out_dir, "artist_tags.csv")
genres_csv  = os.path.join(out_dir, "artist_genres.csv")
isnis_csv   = os.path.join(out_dir, "artist_isnis.csv")
ipis_csv    = os.path.join(out_dir, "artist_ipis.csv")


def export_list_of_dicts(
    df: pd.DataFrame,
    mbid_col: str,
    list_col: str,
    out_csv: str,
    keep_cols=None,
    rename_cols=None,
    first: bool = True
) -> bool:
    """
    Export a one-to-many list-of-dicts field into a child CSV.

    Robust rule here: KEEP first (using original names), then RENAME.
    This avoids losing renamed columns due to keep_cols filtering.
    """
    if list_col not in df.columns or mbid_col not in df.columns:
        return first

    tmp = df[[mbid_col, list_col]].dropna(subset=[list_col])
    if tmp.empty:
        return first

    exploded = tmp.explode(list_col).dropna(subset=[list_col])

    # normalize dicts into columns
    norm = pd.json_normalize(exploded[list_col])
    out = pd.concat([exploded[[mbid_col]].reset_index(drop=True),
                     norm.reset_index(drop=True)], axis=1)

    # keep first (original names)
    if keep_cols:
        keep = [c for c in keep_cols if c in out.columns]
        out = out[keep]

    # then rename
    if rename_cols:
        out = out.rename(columns=rename_cols)

    out.to_csv(out_csv, index=False, mode="w" if first else "a", header=first)
    return False


def export_list_of_strings(
    df: pd.DataFrame,
    mbid_col: str,
    list_col: str,
    out_csv: str,
    value_name: str,
    first: bool = True
) -> bool:
    """Export a one-to-many list-of-strings field into a child CSV."""
    if list_col not in df.columns or mbid_col not in df.columns:
        return first

    tmp = df[[mbid_col, list_col]].dropna(subset=[list_col])
    if tmp.empty:
        return first

    out = tmp.explode(list_col).dropna(subset=[list_col]).rename(columns={list_col: value_name})
    out.to_csv(out_csv, index=False, mode="w" if first else "a", header=first)
    return False


# delete old outputs if re-running
for p in [aliases_csv, tags_csv, genres_csv, isnis_csv, ipis_csv]:
    if os.path.exists(p):
        os.remove(p)

first_aliases = first_tags = first_genres = first_isni = first_ipi = True

chunk_size = 100_000
for i, df in enumerate(pd.read_json(file_path, lines=True, chunksize=chunk_size)):
    df = df.rename(columns={"id": "mbid"})

    # aliases (list of dicts)
    first_aliases = export_list_of_dicts(
        df, "mbid", "aliases", aliases_csv,
        keep_cols=["mbid", "name", "sort-name", "locale", "type", "primary", "begin-date", "end-date", "ended"],
        rename_cols={"name": "alias_name", "sort-name": "alias_sort_name"},
        first=first_aliases
    )

    # tags (list of dicts)  -> will keep 'name' then rename to 'tag'
    first_tags = export_list_of_dicts(
        df, "mbid", "tags", tags_csv,
        keep_cols=["mbid", "name", "count"],
        rename_cols={"name": "tag"},
        first=first_tags
    )

    # genres (list of dicts) -> will keep 'name' then rename to 'genre'
    first_genres = export_list_of_dicts(
        df, "mbid", "genres", genres_csv,
        keep_cols=["mbid", "name", "count", "disambiguation"],
        rename_cols={"name": "genre"},
        first=first_genres
    )

    # isnis (list of strings)
    first_isni = export_list_of_strings(
        df, "mbid", "isnis", isnis_csv, value_name="isni", first=first_isni
    )

    # ipis (list of strings)
    first_ipi = export_list_of_strings(
        df, "mbid", "ipis", ipis_csv, value_name="ipi", first=first_ipi
    )

    if i % 10 == 0:
        print("processed chunk", i)

print("done. outputs in:", out_dir)
```
